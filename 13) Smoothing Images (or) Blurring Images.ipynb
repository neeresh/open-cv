```python
# most common use is to remove noise in the images
# we can also use linear filters , easy to achieve and also relatively fast
# various filters are homogenous filters, gaussian filter, median filter, bilateral filter, etc

# Image smoothing is a key technology of image enhancement, which can remove noise in images.
# Image smoothing is a method of improving the quality of images.
```

```python
# homogenous filter is most simple filter, each output pixel is the mean of its kernel neigbors
# In homogenous filter all pixels contribut equal weight so it is called as homogenous filter
```

```python
# Kernel :-  It is a shape which we can apply or convolve over an image.
```

```python
# homogenoues kernel (k) = 1 / (k_widht * k_height) * ones(width X height)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('opencv-logo.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)

kernel = np.ones((5, 5), np.float32) / 25 # according to formula
destination_image = cv2.filter2D(image, -1, kernel)

titles = ['Image', '2D convolution']
images = [image, destination_image]

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# As in one-dimensional signals, images can also be filtered with the various low-pass filters (LPF) and
# high-pass filters (HPF), etc.

# LPF - helps in eliminating the noise or blurring
# HPD - helps in finding the edges in the images
```

```python
image = cv2.imread('opencv-logo.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32) / 25 # according to formula
destination_image = cv2.filter2D(image, -1, kernel)
blur = cv2.blur(image, (5, 5))

titles = ['Image', '2D convolution', 'Blur']
images = [image, destination_image, blur]

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Gaussians Filter Algorithm :- 
# It is nothing but using different weight kernel, in both x and y direction
```

```python
image = cv2.imread('opencv-logo.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32) / 25 # according to formula
destination_image = cv2.filter2D(image, -1, kernel)
blur = cv2.blur(image, (5, 5))
Gblur = cv2.GaussianBlur(image, (5, 5), 0)

titles = ['Image', '2D convolution', 'Blur', 'Gaussian Blur']
images = [image, destination_image, blur, Gblur]

for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Median Filter :- It is something that replaces each pixels value with the median of its neighbouring pixels.
# This method is great when dealing with salt and pepper noise.
```

```python
image = cv2.imread('salt_pepper.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32) / 25 # according to formula
destination_image = cv2.filter2D(image, -1, kernel)
blur = cv2.blur(image, (5, 5))
Gblur = cv2.GaussianBlur(image, (5, 5), 0)
median = cv2.medianBlur(image, 5) # kernel size should only be odd except 1.

titles = ['Image', '2D convolution', 'Blur', 'Gaussian Blur', 'Median']
images = [image, destination_image, blur, Gblur, median]

for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# HPF - Helps finding the edges in the images
# Bilateral Filter
```

```python
image = cv2.imread('lena.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32) / 25 # according to formula
destination_image = cv2.filter2D(image, -1, kernel)
blur = cv2.blur(image, (5, 5))
Gblur = cv2.GaussianBlur(image, (5, 5), 0)
median = cv2.medianBlur(image, 5) # kernel size should only be odd except 1.
bilateral = cv2.bilateralFilter(image, 9, 75, 75) #use google to find more about syntax

titles = ['Image', '2D convolution', 'Blur', 'Gaussian Blur', 'Median', 'Bilateral Filter']
images = [image, destination_image, blur, Gblur, median, bilateral]

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
