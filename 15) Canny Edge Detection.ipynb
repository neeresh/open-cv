```python
# Canny Edge detector is an edge detection operator that uses multi-stage algorithm
# to detect a wide range of edges in images.

# Canny edge detection algorithm is performed in 5 steps:-
# 1) Noise reduction :- To apply Gaussiann filter to smooth the image and to eliminate noise
# 2) Gradient Calculation :- Find the intensity gradients of the image
# 3) Non-maximum suppression :- To apply non-max suppression to get rid of spuriors respones to edge detection.
# 4) Double Threshold :- To apply double threshold to determine to determine potential edges.
# 5) Edge Tracking by hysteresis :- To track edges by hysteresis. i.e, To finalize the detection of the edges by suppresing 
# all the other edges that are weak that are not connected to all the edges.
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('messi5.jpg', 0)

# th1 and th2 is for hysteresis procedure

canny = cv2.Canny(image, 100, 200)

titles = ['Normal_Image', 'Canny Edge Detection']
images = [image, canny]

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# combining sobelX and sobelY

image = cv2.imread('messi5.jpg', 0)

# Laplacian method
lap = cv2.Laplacian(image, cv2.CV_64F, ksize = 1)
lap = np.uint8(np.absolute(lap)) 

# Sobel X method
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0) # dx=1 for x direction and dy=0 is for y direction
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1) 


# we are going to convert these values to unsign int like in case of laplacian
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX, sobelY)

# Canny edge detection
canny = cv2.Canny(image, 100, 200)

titles = ['Normal_Gray image', 'Laplacian', 'SobelX', 'SobelY', 'Sobel Combined', 'Canny Edge Detection']
images = [image, lap, sobelX, sobelY, sobelCombined, canny]

for i in range(6):
    plt.subplot(2 , 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])
plt.show()
```
