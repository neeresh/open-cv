```python
# Morphological Transformations are simple operations based on the image shape
# it is normally performed on binary image and when we performed morphological transformation there are 2 things required
# 1st is original image 2nd is structuring element or kernel which decides the nature of operation.

# Morphology is a broad set of image processing operations that process images based on shapes. 
# Morphological operations apply a structuring element to an input image, creating an output image of the same size.

# *** Morphological Image Processing. Binary images may contain numerous imperfections.
# In particular, the binary regions produced by simple thresholding are distorted by noise and texture.
# Morphological image processing pursues the goals of removing these imperfections by accounting for 
# the form and structure of the image.***
```

 ## MORPHOLOGICAL MEHTODS:-

```python
# Normally we perform Morphological Transformation on binary images

# Dilation

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((2,2), np.uint8)
dilation = cv2.dilate(mask, kernal)

titles = ['Image', 'Mask', 'Dilation']
images = [image,mask,dilation]

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# To remove more black dots on dilation image, use iterations in cv2.dilate()

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((2,2), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 2)

titles = ['Image', 'Mask', 'Dilation']
images = [image,mask,dilation]

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# we can change np.ones() to get clear transformation but it is not mostly recommended it may change the image

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((5,5), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 2)

titles = ['Image', 'Mask', 'Dilation']
images = [image,mask,dilation]

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Erosion :- It erodes away the budary of the fore ground object . so, when erosion is applied the kernal which we defined 
# slides through all the image and a pixel in original image either 1 or 0 will be considered 1 if all the pixels 
# under the kernal is 1 otherwise it is eroded i.e, value is 0 ( black area).

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((2, 2), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 2)
erosion = cv2.erode(mask, kernal ,iterations = 2)

titles = ['Image', 'Mask', 'Dilation', 'Erosion']
images = [image,mask,dilation,erosion]

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Opening :- another name of "erosion" followed by "dialation" i.e, first erosion if performed on image 
# then the dialation is performed on the image.

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((2, 2), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 2)
erosion = cv2.erode(mask, kernal ,iterations = 2)
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)

titles = ['Image', 'Mask', 'Dilation', 'Erosion', 'Opening']
images = [image, mask, dilation, erosion, opening]

for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Closing :- another name of "dialation" followed by "erosion" i.e, first dialation if performed on image 
# then the erosion is performed on the image.

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((2, 2), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 2)
erosion = cv2.erode(mask, kernal ,iterations = 2)
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)

titles = ['Image', 'Mask', 'Dilation', 'Erosion', 'Opening', 'Closing']
images = [image, mask, dilation, erosion, opening, closing]

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
 # different types of morphological operations :-

image = cv2.imread('smarties.png', 0)
_, mask = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY_INV)
# To remove black dots on the balls in the image that is coverted to binary we nneed to use cv2.dilate()

kernal = np.ones((2, 2), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 2)
erosion = cv2.erode(mask, kernal ,iterations = 2)
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)
# morphological gradient
mg = cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernal)
# top hat 
th = cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernal)

titles = ['Image', 'Mask', 'Dilation', 'Erosion', 'Opening', 'Closing', 'Mg', 'Th']
images = [image, mask, dilation, erosion, opening, closing, mg, th]

for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
