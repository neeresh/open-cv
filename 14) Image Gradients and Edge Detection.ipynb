```python
# An image gradient is a directional change in the intensity or color in an image.
# Fundamental building blocks in the image processing.

# Laplacian derivatives, Sobel X methods, Sobel Y methods are different gradient functions which uses different mathmetical
# operations to produce the req image.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('messi5.jpg', 0)

titles = ['Normal_Gray image']
images = [image]

for i in range(1):
    plt.subplot(1, 1, i+1)
    plt.imshow(images[i], 'gray')
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
image = cv2.imread('messi5.jpg', 0)

# Laplacian method
lap = cv2.Laplacian(image, cv2.CV_64F, ksize = 1)
# cv2.CV_64F is a datatype , we are using a 64 bit float due to negative slope induced by 
# transforming the image from white to black.
# It supports the negative numbers which we will dealing with when laplacian method is run on our image.
lap = np.uint8(np.absolute(lap)) # we take abs value of laplacian image tranformation
                                 # and convert this value to unsingned 8 bit integer which is suitable
                                 # to our output



titles = ['Normal_Gray image', 'Laplacian']
images = [image, lap]

for i in range(2):
    plt.subplot(1 , 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
image = cv2.imread('messi5.jpg', 0)

# Laplacian method
lap = cv2.Laplacian(image, cv2.CV_64F, ksize = 1)
lap = np.uint8(np.absolute(lap)) 

# Sobel X method
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 1) # dx=1 for x direction and dy=0 is for y direction
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 1) 

# we are going to convert these values to unsign int like in case of laplacian
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

titles = ['Normal_Gray image', 'Laplacian', 'SobelX', 'SobelY']
images = [image, lap, sobelX, sobelY]

for i in range(4):
    plt.subplot(2 , 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
image = cv2.imread('sudoku.png', 0)

# Laplacian method
lap = cv2.Laplacian(image, cv2.CV_64F, ksize = 1)
lap = np.uint8(np.absolute(lap)) 

# Sobel X method
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 1) # dx=1 for x direction and dy=0 is for y direction
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 1) 

# we are going to convert these values to unsign int like in case of laplacian
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

titles = ['Normal_Gray image', 'Laplacian', 'SobelX', 'SobelY']
images = [image, lap, sobelX, sobelY]

for i in range(4):
    plt.subplot(2 , 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])
plt.show()
```

```python
# combining sobelX and sobelY

image = cv2.imread('sudoku.png', 0)

# Laplacian method
lap = cv2.Laplacian(image, cv2.CV_64F, ksize = 1)
lap = np.uint8(np.absolute(lap)) 

# Sobel X method
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0) # dx=1 for x direction and dy=0 is for y direction
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1) 


# we are going to convert these values to unsign int like in case of laplacian
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX, sobelY)

titles = ['Normal_Gray image', 'Laplacian', 'SobelX', 'SobelY', 'Sobel Combined']
images = [image, lap, sobelX, sobelY, sobelCombined]

for i in range(5):
    plt.subplot(2 , 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])
plt.show()
```
